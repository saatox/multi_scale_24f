```python
import pandas as pd
import numpy as  np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
og_file=pd.read_csv(f'spotify_playlists/liked_songs.csv', header=0)
og_file.head()
```

```python
plt.figure()
og_file['Tempo'].plot.kde()
plt.xticks(fontsize=4)
plt.tight_layout()
```

```python
plt.figure()
og_file['Popularity'].plot.kde()
plt.xticks(fontsize=4)
plt.tight_layout()
```

```python
artists=pd.DataFrame()
artists['Artists']=og_file['Artist Name(s)'].value_counts().index.to_list()
artists['Count']=og_file['Artist Name(s)'].value_counts().to_list()
```

```python
plt.figure()
artists[:30].plot.bar(y='Count')
artist_list=artists['Artists'].to_list()
artist_list=artist_list[:30]
print(artist_list)
plt.xticks(ticks=range(0, 30),labels=artist_list)
plt.tight_layout()
```

```python
this_year=pd.DataFrame()
this_year=og_file[og_file['Added At'].str.contains('2024', na=False)]
this_year.head()
```

```python
artists=pd.DataFrame()
artists['Artists']=this_year['Artist Name(s)'].value_counts().index.to_list()
artists['Count']=this_year['Artist Name(s)'].value_counts().to_list()
print(artists.head())
```

```python
plt.figure()
artists[:30].plot.bar(y='Count')
artist_list=artists['Artists'].to_list()
artist_list=artist_list[:30]
print(artist_list)
plt.xticks(ticks=range(0, 30),labels=artist_list)
plt.tight_layout()
```
