```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
temp=[]
with open('results.txt') as f:
    for line in f:
        line=line.strip()
        line=line.replace("[", "")
        line=line.replace("]", "")
        line=line.split(",")
        for i in range(len(line)):
            line[i]=float(line[i])
        temp.append(line)

print(temp)
```

```python
plt.plot(temp)
plt.show()
```
