```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
temp=[]
with open('results.txt') as f:
    for line in f:
        line=line.strip()
        line=line.replace("[", "")
        line=line.replace("]", "")
        line=line.split(",")
        for i in range(len(line)):
            line[i]=float(line[i])
        temp.append(line)

print(temp)
```

```python
temp_x=[]
for i in range(0, len(temp)):
    temp_x.append(temp[i][0])
temp_y=[]
for i in range(0, len(temp)):
    temp_y.append(temp[i][1])
temp_z=[]
for i in range(0, len(temp)):
    temp_z.append(temp[i][2])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#ax.plot_trisurf(temp_x, temp_y, temp_z, color='white', edgecolors='grey', alpha=0.5)
ax.scatter(temp_x, temp_y, temp_z, c='red')
plt.show()
```
